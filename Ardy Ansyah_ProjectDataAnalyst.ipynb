## Data Manipulation

### Total Transaction by Customer

```python
total_transaksi = df.groupby('CUSTOMERNAME')['ORDERNUMBER'].nunique().reset_index()
total_transaksi.columns = ['CUSTOMERNAME', 'TOTAL_TRANSACTIONS']

top10 = total_transaksi.sort_values(by='TOTAL_TRANSACTIONS', ascending=False).head(10)

plt.figure(figsize=(12, 6))

sns.barplot(data=top10, x='TOTAL_TRANSACTIONS', y='CUSTOMERNAME', palette='pastel')

plt.title('Top 10 Pelanggan Berdasarkan Jumlah Transaksi', fontsize=14)
plt.xlabel('')
plt.ylabel('')

for index, value in enumerate(top10['TOTAL_TRANSACTIONS']):
    plt.text(value + 0.2, index, str(value), va='center')

plt.tight_layout()
plt.show()
```
